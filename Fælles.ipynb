```python
import pandas as pd
import ast
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("/Users/MathildeStouby/Untitled Folder/tmdb_5000_movies.csv")
```

```python
subset=df[['title','budget','revenue','genres','keywords','popularity','vote_average','vote_count','production_countries','release_date']]
```

```python
subset=subset[subset.budget != 0]
subset=subset[subset.revenue != 0]
```

```python
subset['overskud i procent'] = (subset.revenue/subset.budget)*100
```

```python
countries=[]

for liste in subset['production_countries']:
    Temp = ast.literal_eval(liste)
    countries.append(Temp)

onlycountries=[]

for i in range(len(countries)):
    templist=[]
    for j in range(len(countries[i])):
        x = countries[i][j]["name"]
        templist.append(x)
    onlycountries.append(templist)
```

```python
subset['produktionsland']=onlycountries
subset=subset.drop('production_countries',axis=1)
```

```python
keywords=[]

for liste in subset['keywords']:
    Temp = ast.literal_eval(liste)
    keywords.append(Temp)

onlykeywords=[]

for i in range(len(keywords)):
    templist=[]
    for j in range(len(keywords[i])):
        x = keywords[i][j]['name']
        templist.append(x)
    onlykeywords.append(templist)
```

```python
subset['nøgleord']=onlykeywords
subset=subset.drop('keywords',axis=1)
```

```python
genres=[]

for liste in subset['genres']:
    Temp = ast.literal_eval(liste)
    genres.append(Temp)

onlygenres=[]

for i in range(len(genres)):
    templist=[]
    for j in range(len(genres[i])):
        x = genres[i][j]['name']
        templist.append(x)
    onlygenres.append(templist)
```

```python
subset['genrer']=onlygenres
subset=subset.drop('genres',axis=1)
```

```python
yes_or_no_USA=[]

for liste in subset.produktionsland:
    if "United States of America" in liste:
        yes_or_no_USA.append(1)
    else:
        yes_or_no_USA.append(0)

subset['USA']=yes_or_no_USA
```

```python
subset=subset[subset.USA == 1]
subset=subset.drop('USA',axis=1)
```

```python
listofgenres=['Drama',
 'Comedy',
 'Thriller',
 'Action',
 'Romance',
 'Adventure',
 'Crime',
 'Science Fiction',
 'Horror',
 'Family',
 'Fantasy',
 'Mystery',
 'Animation',
 'History',
 'Music',
 'War',
 'Documentary',
 'Western',
 'Foreign',
 'TV Movie']
```

```python
i=0

while i<len(listofgenres):
    temp=[]
    for liste in subset.genrer:
        if listofgenres[i] in liste:
            temp.append(1)
        else:
            temp.append(0)
    subset[str(listofgenres[i])]=temp
    i+=1
    
```
